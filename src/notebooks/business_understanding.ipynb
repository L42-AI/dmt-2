## [0]
- Task: Classification of user preferences into hotel 'clusters' grouped by Expedia
- Features used:
    - Temporal:
        - Generated day, month, and year features from timestamps of search and check-in/out (month turns out to be more significant predictor). 
        - Length of stay, shorter or longer stays are more suitable for certain hotels 
    - Geographical:
        - Author claims distance between user and hotel is significant predictor, did log(distance) to tune down large distances, and square(distance) to empathize large distances more.
    - IDs
        - Author dummy encoded IDs into binary features, which led to many features and sparsity. To alleviate this, author did feature selection of features that most impact test scores. 
    - Latent features from reviews
        - Author used PCA to reduce latent variables to 2 components
