# Business Understanding
Your task is to predict what hotel properties that result from a search of a user, the user is
most likely to click on. Of course, more people have worked on such predictions. Can you
find some other people that have tried to make such predictions (e.g. from the Kaggle com-
petition)? And what have they used as most prominent predictors? Have other people that
participate in the competition mentioned anything about their approaches? Please spend a
couple of paragraphs on this topic in a ’Related Work’ section in your report.

## [0] https://cs229.stanford.edu/proj2016spr/report/022.pdf
- Task: Classification of user preferences into hotel 'clusters' grouped by Expedia
- Processing:
    - Temporal:
        - Generated day, month, and year features from timestamps of search and check-in/out (month turns out to be more significant predictor). 
        - Length of stay, shorter or longer stays are more suitable for certain hotels 
    - Geographical:
        - Author claims distance between user and hotel is significant predictor, did log(distance) to tune down large distances, and square(distance) to empathize large distances more.
    - IDs
        - Author dummy encoded IDs into binary features, which led to many features and sparsity. To alleviate this, author did feature selection of features that most impact test scores. 
    - Latent features from reviews
        - Author used PCA to reduce latent variables to 2 components

## [1] https://arxiv.org/pdf/1311.7679
The authors placed 5th in the ICDM 2013 challenge.
- Task: Same as for us, ranking hotels according to their likelihood of being selected ("rank hotels to maximize purchases for given hotel queries")
- Processing:
    - The authors do not share much information on the specific features, only general methodology
    - Feature ranks
        - "Some of the feature's rank in the list as single feature for each hotel choice"
        - Not sure what this means but I think: For each query, encode the feature's rank in the query as a new feature
        - So, if in a single search, hotel #4 has the cheapest price, it has rank 1.
    - Composite features
        - For example: srch_room_count * max(srch_booking_window) + srch_booking_window
        - Claude: The author likely believed these two features capture a combined demand/intent signal that neither captures alone:
            Room count reflects the size of the travel group — solo traveler vs. family vs. corporate group
            Booking window reflects planning horizon — last-minute vs. far-in-advance
            Together, they describe a very specific type of traveler intent:

            A large group booking far in advance → likely a planned event (wedding, conference), very price-sensitive, high commitment

            A single room booked last minute → business traveler or spontaneous trip, potentially less price-sensitive

            A large group booking last minute → unusual, possibly signals an emergency or special circumstance
    - Bucket/binning float features:
        - developers.google.com: Binning is a good alternative to scaling or clipping when either of the following conditions is met: 
            - The overall linear relationship between the feature and the label is weak or nonexistent.
            - When the feature values are clustered.
            - They recommend quantile bucketing for skewed data. Equal intervals empathize information in the long tail, quantile bucketing empathizes information in the torso/body of the data.
