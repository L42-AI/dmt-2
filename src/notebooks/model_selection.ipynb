## Model Selection Options

| Model | Type | Main idea | Strength | Weakness | Priority |
|---|---|---|---|---|---|
| Rule-based ranking | Baseline | Rank hotels using manually chosen features such as review score, location score, promotion flag and price rank | Easy to explain, useful benchmark | Too simple, no learning | High |
| Logistic Regression / sklearn model | Pointwise ML baseline | Predict click, booking or relevance per row, then sort within search | Easy to implement and explain | Does not directly optimize ranking | High |
| Random Forest / Gradient Boosting | Pointwise ML baseline | Predict relevance score using nonlinear tabular model | Captures interactions better than logistic regression | Can be slow on 5 million rows | Medium |
| XGBoost Ranker | Learning-to-rank | Optimizes ranking objective such as NDCG | Strong fit for competition metric | Needs grouped data and tuning | High |
| LightGBM Ranker / LambdaMART | Learning-to-rank | Learns ranking within each search using LambdaRank/LambdaMART-style objective | Fast and strong for large tabular ranking problems | More complex than sklearn | Very High |
| PyTorch Neural Ranker | Deep learning | Train neural network or pairwise ranking model | Flexible, can use embeddings | More work, risky under time pressure | Optional |

### Model 0: Rule-Based Baseline

Before training complex models, we should create a simple rule-based ranking baseline.

Possible score:

- Higher review score is better
- Higher location score is better
- Promotion flag may increase attractiveness
- Lower price rank within the same search may be better
- Branded hotels may perform differently from independent hotels

This baseline helps us check whether machine learning actually improves performance.

### Model 1: sklearn Pointwise Baseline

A simple machine learning baseline can predict the relevance score for each row independently.

Possible models:

- Logistic Regression
- Random Forest
- HistGradientBoosting
- GradientBoostingRegressor

The predicted score is then used to rank properties within each `srch_id`.

This is not a perfect match for the task, because the model does not directly learn rankings. However, it is useful as a simple baseline.

### Model 2: LambdaMART / LightGBM Ranker

This is likely our strongest main candidate.

LambdaMART is a learning-to-rank method based on gradient-boosted decision trees.  
It is designed to learn the correct order of items within a query group.

For this assignment, each `srch_id` is a query group.  
The model learns to rank hotels inside each search.

This matches the Kaggle objective better than standard classification or regression, because the competition evaluates the ranked list using NDCG@5.

### Model 3: XGBoost Ranker

XGBoost Ranker is another strong candidate.

It is similar in spirit to LightGBM Ranker: it uses boosted decision trees but optimizes a ranking objective.

Possible objective:

- `rank:ndcg`

This model is worth trying if LightGBM is unstable, unavailable, or if we want to compare two ranking models.

### Model 4: PyTorch Neural Ranking Model

A PyTorch model could be used to build a neural ranking system.

Possible ideas:

- MLP predicting relevance
- Pairwise RankNet-style model
- Hotel and destination embeddings
- Wide-and-deep architecture

However, this is probably not the safest first choice because:

- It requires more custom code.
- It is harder to debug.
- It may need more tuning.
- Tree-based models usually perform very well on tabular data.

Recommendation: only try PyTorch if the LightGBM or XGBoost ranking pipeline is already working.
